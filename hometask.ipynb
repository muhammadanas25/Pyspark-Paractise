```
from pyspark.sql import SparkSession 

spark=SparkSession.builder.master('local[*]').appName('my-app').getOrCreate()
spark
```

```
data=spark.read.format('csv').option('inferschema','true').option('header','true').load('/FileStore/tables/employee_earnings_report_2016.csv')
```

```
data.show()
```

```
data.groupby(['Gender','Title']).count().show(truncate=False)
```

```
from pyspark.sql.functions import col, asc,desc
data.groupby('POSTAL').count().orderBy(col('count'),ascending=False).limit(3).show(truncate=False)
```

```
data.groupby([ 'DEPARTMENT_NAME','Title']).count().show(truncate=False)
```

```
data.createOrReplaceTempView("people")

sqlDF = spark.sql('''SELECT GENDER,Title,count(*) FROM people group by GENDER,Title''')
sqlDF.show(truncate=0)
```

```
data.createOrReplaceTempView("people")

sqlDF = spark.sql('''SELECT POSTAL, count(POSTAL)as count FROM people group by POSTAL 
order by count desc LIMIT 3
''')
sqlDF.show(truncate=0)
```

```
sqlDF = spark.sql('''SELECT DEPARTMENT_NAME, Title FROM people group by DEPARTMENT_NAME,Title

''')
sqlDF.show(truncate=0)
```
